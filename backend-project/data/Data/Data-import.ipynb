```python
import pandas as pd
import os
from IPython import display
import re
from sqlalchemy import create_engine
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Extract League and Season information from filename
regex = r".+\. (.+) \- (\d{4}-\d{4}).xls$"
frames = []
s = 0
for file in os.listdir("."):
    filename = os.fsdecode(file)
    m = re.match(regex, filename)
    if m is not None:
        s += 1
        xl = pd.ExcelFile(filename)
        # Select the right sheet
        df = xl.parse("Box score")
        # Clean data
        # Rename first three columns
        df.insert(len(df.columns), "Season", [m.group(2)] * len(df), False)
        df.insert(len(df.columns), "League", [m.group(1)] * len(df), False)
        print(f"Processed {m.group(1)} - {m.group(2)}")
        frames.append(df)

df = pd.concat(frames, ignore_index=True)

# Clean up the data
df = df.replace("-", "0")
df = df.rename(columns={"Unnamed: 0": "Jersey number", "Unnamed: 1": "Player name", "Unnamed: 2": "Team name"})
df["Points"] = df["Points"].astype(int)
```

```python
df[df["League"] == "Liga Endesa"].sort_values("Points", ascending=False)
```

```python
engine = create_engine("sqlite://")
df.to_sql("Players", con=engine)
```
